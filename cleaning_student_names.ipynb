```python
# Data manipulation: picking out and removing prefixes and suffixes

import os
import pandas as pd

schools_data = os.path.join("Resources", "schools_complete.csv")
students_data = os.path.join("Resources", "students_complete.csv")
missing_data = os.path.join("Resources", "missing_grades.csv")

schools_df = pd.read_csv(schools_data)
students_df = pd.read_csv(students_data)
missing_df = pd.read_csv(missing_data)

#Fetch only one column student name
students_names = students_df["student_name"].tolist()

#Loop through student names list and split it then print
name_audit_list = []
for name_split in students_names:
    name = name_split.split()
    if len(name) >= 3:
        name_audit_list.append(name_split)
        
# Extract prefixes and put in list
prefixes = []

for name2 in name_audit_list:
    if len(name2.split()[0]) <= 4:
        prefixes.append(name2.split()[0])
        
suffixes = []
for name3 in name_audit_list:
    if len(name3.split()[-1]) <= 3:
        suffixes.append(name3.split()[-1])
        
# Using .set() method to return all unique items in a list 
# Truncated dataset easy to pick out things that can be excluded.

print(f'{set(prefixes)}\n')
print(f'{set(suffixes)}')

prefix_suffixes = ['Miss', 'Ms.', 'Mrs.','Mr.', 'Dr.', 'DVM', 'MD', 'PhD', 'DDS']

for txt_remove in prefix_suffixes:
        students_df["student_name"] = students_df["student_name"].str.replace(txt_remove,"")
                    
students_df.head(10)
```

```python
# bin all schools spending range, and give the average: math score, reading score, passing math rate, passing reading rate, overall passing rate

math_bin = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["Average Math Score"]
reading_bin = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["Average Reading Score"]
math_pass_bin = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["% Passing Math"]
reading_pass_bin = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["% Passing Reading"]
overall_pass_bin = per_school_summary_df.groupby(["Spending Range (Per Student)"]).mean()["% Overall Passing"]

overall_pass_bin
```
